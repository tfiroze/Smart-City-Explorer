```python
# Encapsulation: determine whether a given point is in a given region

def match_venues_zones(pos, point):
    df = pd.read_csv("taxi_zones.csv")

    # Create Shapely's Point object
    point = Point(point[0], point[1])

    # Create Shapely's MultiPolygon object
    multipolygon = loads(pos)

    # Check if the point is in the polygon
    res =  point.within(multipolygon)

    return res
```

```python
# Iterate over each row of the_geom column in the data set of all taxis zones

import csv

zone_arr = []
zone_id = []

with open('taxi_zones.csv', 'r') as file:

    reader = csv.reader(file)
    
    # Read header row
    header = next(reader)
    
    # Determine the index of the column to be traversed
    target_column_index = header.index('the_geom')
    target_column_id = header.index('OBJECTID')
    
    # Iterate through each row
    for row in reader:
        # Get the value of the specified column and add it to the array
        column_value = row[target_column_index]
        zone_arr.append(column_value)

        # Store the zone id info into another array 
        column_value_id = row[target_column_id]
        zone_id.append(column_value_id)
```

```python
# Define the data to be written
data_to_append = venue_arr

# Open CSV file to write data in append mode
with open('venues_in_zones.csv', 'a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write to data row
    for row in data_to_append:
        writer.writerow(row)

print("The data has been successfully appended to the CSV file")
```
